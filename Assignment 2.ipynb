group the profit as low and high profit.


```python
%%sql
-- Query 1

Select city, customer_type, unit_price, quantity, total, cogs
from sales
```

```python
%%sql
-- Query 2

select (total - cogs) AS profit
from (Select city, customer_type, unit_price, quantity, total, cogs
                                    from sales) sub
order by profit desc
```

```python
%%sql
-- Query 3

select city, gender, product_line, rating
from sales
```

```python
%%sql
--Query 4
select
    case 
        when rating >= 8 then 'High'
        when rating >= 6 then 'Medium'
        ELSE 'Low'
    END rating
from (select city, gender, product_line, rating
                            from sales) sub1
```

```python
%%sql
-- Query 5
select 
    CASE
        when profit >=24.8 then 'High'
        ELSE 'Low'
    END profit
from (select (total - cogs) AS profit
                    from (Select city, customer_type, unit_price, quantity, total, cogs
                                    from sales) sub) sub_2
```
