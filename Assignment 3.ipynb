```python
%%sql
-- Query 1
-- Display countries and their population size.

SELECT 
     c.name country,
     p.size as population 
FROM 
    countries c
INNER JOIN populations p
    ON c.code = p.country_code
LIMIT 10
```

```python
%%sql
-- Query 2
-- display countries and their currencies

SELECT
    c.name country,
    cu.basic_unit currencies
FROM 
    countries c
INNER JOIN currencies cu
    USING(code)
limit 10
```

```python
%%sql
-- Query 3
-- display countries and their fertility rate

SELECT c.name, p.fertility_rate 
FROM countries c
    INNER JOIN populations p
    ON c.code = p.country_code
LIMIT 10
```

```python
%%sql
-- Query 4
-- display countries, continent, income group, and thier total investment

SELECT c.name "Country", c.continent "Continent", e.income_group "Income Group", e.total_investment "Total Investment"
FROM countries c
    JOIN economies e
    USING(code)
LIMIT 10
```

```python
%%sql
-- Query 5
-- display cities, country, region, independence year

SELECT ci.name city, c.name country, c.region region, c.indep_year Independence_year
FROM countries c
    JOIN cities ci
    ON c.code = ci.country_code
limit 10
```

```python
%%sql
-- Query 6
-- display Asian countries, and their langauges

SELECT c.name Asian_countries, l.name languages
FROM countries c
    INNER JOIN languages l
    USING(code)
WHERE c.continent = 'Asia'
limit 10
```

```python
%%sql
-- Query 7
-- display European countries, life expectancy, inflation rate

SELECT 
    c.name "European Countries", 
    p.life_expectancy "Life Expectancy", 
    e.inflation_rate "Inflation Rate"
FROM
    countries c
INNER JOIN populations p
    ON c.code = p.country_code
INNER JOIN economies e
    USING(code)
WHERE 
    c.continent = 'Europe'
LIMIT 10
```
